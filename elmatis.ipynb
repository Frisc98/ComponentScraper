```python
import requests
from bs4 import BeautifulSoup
```

```python
URL_base = "http://www.elmatis.hr/Articles.aspx"
page = requests.get(URL_base)
soup = BeautifulSoup(page.content, "html.parser")
category_list = soup.find("div", id="navigationContainer").find_all("a")
```

```python
print(category_list)
```

```python
category = "I. C. KRUGOVI I PODNOŽJA"
category_links = []
N = len(category_list)
found_subpage = False
for i_cat in range(N):
    subpage_addr = category_list[i_cat].get("href")
    if (subpage_addr == "#"):
        if (found_subpage):
            break
        subpage_name = category_list[i_cat].text.strip()
        if (subpage_name == category):
            found_subpage = True
    elif (found_subpage):
        category_links.append(subpage_addr)
if (len(category_links) == 0):
    print("no such category")
print(category_links)
```

```python
URL_base_2 = "http://www.elmatis.hr"
URL_i = URL_base_2 + category_links[0]
page = requests.get(URL_i)
soup = BeautifulSoup(page.content, "html.parser")
print(soup)
```

```python
results = soup.find("table", class_="list")
component_PN = []
component_desc = []
component_price = []
component_stock = []
rows = results.find_all("tr")
Nrows = len(rows)
## last two rows are for page numbers
for i in range(Nrows-2):
    row = rows[i]
    cells = row.find_all("td")
    if (len(cells) > 0):
        component_PN.append(cells[0].text.strip())
        component_desc.append(cells[1].text.strip())
        component_price.append(cells[2].find("div", class_="price-kn").text.strip())
        component_stock.append(cells[3].text.strip())
```

```python
for i in range(len(component_PN)):
    print(f"Part Number: {component_PN[i]}\t\tPrice: {component_price[i]}\t\tDescription: {component_desc[i]}\t\tStock: {component_stock[i]}")
```

```python
VIEWSTATEvalue  = soup.find("input", id="__VIEWSTATE").get("value")
VIEWSTATEGENERATORvalue  = soup.find("input", id="__VIEWSTATEGENERATOR").get("value")
EVENTVALIDATIONvalue  = soup.find("input", id="__EVENTVALIDATION").get("value")
req_page = 9
data = {
    "__EVENTTARGET" : "ctl00$Main$ucProductList$grdProductListPublic",
    "__EVENTARGUMENT" : "Page${}".format(req_page),
    "__VIEWSTATEGENERATOR" : VIEWSTATEGENERATORvalue,
    "__VIEWSTATE" : VIEWSTATEvalue,
    "__EVENTVALIDATION" : EVENTVALIDATIONvalue
}
page2 = requests.post(URL_i, data)
soup2 = BeautifulSoup(page2.content, "html.parser")
print(soup2.find("title").get_text())
#print(soup2.find("table", class_="list"))
```
